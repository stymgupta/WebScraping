**Import Dependencies**

```python
from requests import get
from bs4 import BeautifulSoup as Soup
import pandas as pd
```

**Getting URL**

```python
url = get('https://imdb.to/3CzG7te')
```

***Parsing Data using Soup***

```python
Soup_data = Soup(url.text, "html.parser")
```

```python
#Finding all data using div
Movies = Soup_data.findAll("div",{'class':'lister-item mode-advanced'})
```

**Extracting Data from all**

```python
name = []
time = []
for i in Movies:
  name.append(i.h3.a.text)
  time.append(i.p.find('span',{'class':'runtime'}).text.split()[0])
```

```python
#wrapping all data into list
data = list(zip(name,time))
```

```python
#Creating DataFrame
df = pd.DataFrame(data, columns=['name','time'])
```

```python
#Creating New Column for desired output
df['complete'] = df['name'] + ' ' + df['time']
```

```python
df
```

```python
#Convert time column as number
df['time'] = pd.to_numeric(df.time)
```

```python
t1 = int(input('Enter First Value: '))
t2 = int(input('Enter Second Value: '))
```

**Creating List and then Comvert into Dictionary to access output as a key value pair**

```python
values=[]
runtime = []
for i in range (t1-1,t2):
  values.append(df.loc[i]['complete'])
  runtime.append(df.loc[i]['time'])
```

```python
com_dict = {runtime[i]: values[i] for i in range(len(runtime))}
com_dict = dict(sorted(com_dict.items()))
print(com_dict.get(list(com_dict)[-1]))
```
